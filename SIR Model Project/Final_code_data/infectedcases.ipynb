### Creating one graph with all states - infected cases and deaths

```python
#importing packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
import io
```

```python
#download the data
url = "https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv"
download = requests.get(url).content
df = pd.read_csv(io.StringIO(download.decode('utf-8')))

state1 = "Maine"
state2 = "Washington"
state3 = "Texas"
state4 = "Nebraska"
state5 = "California"

#subset Maine data & reset index
df_state1 = df[df.state == state1]
df_state2 = df[df.state == state2]
df_state3 = df[df.state == state3]
df_state4 = df[df.state == state4]
df_state5 = df[df.state == state5]

```

```python
df_state1.reset_index(inplace=True,drop=True)


x = list(df_state1["cases"])
for i in range(0,13):
    x.insert(0,0)
x = x[:-13]
df_state1["infected"] = df_state1["cases"] - x

df_state1
```

```python
df_state2.reset_index(inplace=True,drop=True)


x = list(df_state2["cases"])
for i in range(0,13):
    x.insert(0,0)
x = x[:-13]
df_state2["infected"] = df_state2["cases"] - x

df_state2
```

```python
df_state3.reset_index(inplace=True,drop=True)


x = list(df_state3["cases"])
for i in range(0,13):
    x.insert(0,0)
x = x[:-13]
df_state3["infected"] = df_state3["cases"] - x

df_state3
```

```python
df_state4.reset_index(inplace=True,drop=True)


x = list(df_state4["cases"])
for i in range(0,13):
    x.insert(0,0)
x = x[:-13]
df_state4["infected"] = df_state4["cases"] - x

df_state4
```

```python
df_state5.reset_index(inplace=True,drop=True)


x = list(df_state5["cases"])
for i in range(0,13):
    x.insert(0,0)
x = x[:-13]
df_state5["infected"] = df_state5["cases"] - x

df_state5
```

# Exploratory Data Analysis

```python
#plotting cases

fig = plt.figure(figsize=(15,10))

plt.subplot(1, 2, 1) 
plt.plot(df_state1.index, df_state1['cases'],label='Maine')
plt.plot(df_state2.index, df_state2['cases'],label='Washinton')
plt.plot(df_state3.index, df_state3['cases'],label='Texas')
plt.plot(df_state4.index, df_state4['cases'],label='Nebraska')
plt.plot(df_state5.index, df_state5['cases'], label='Califronia')
plt.xlabel("Time Since First Case (Days)")
plt.ylabel("Count")
plt.title("Cumulative Cases")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(df_state1.index, df_state1['deaths'],label='Maine')
plt.plot(df_state2.index, df_state2['deaths'],label='Washinton')
plt.plot(df_state3.index, df_state3['deaths'],label='Texas')
plt.plot(df_state4.index, df_state4['deaths'],label='Nebraska')
plt.plot(df_state5.index, df_state5['deaths'], label='Califronia')
plt.xlabel("Time Since First Case (Days)")
plt.ylabel("Count")
plt.title("Cumulative Deaths")
plt.legend()
```

```python
fig = plt.figure(figsize=(25,15))
plt.rcParams['font.size'] = '16'
plt.plot(df_state1.index, df_state1['infected']/1362359,label='Maine')
plt.plot(df_state2.index, df_state2['infected']/7766925,label='Washinton')
plt.plot(df_state3.index, df_state3['infected']/29527941,label='Texas')
plt.plot(df_state4.index, df_state4['infected']/7766925,label='Nebraska')
plt.plot(df_state5.index, df_state5['infected']/39664128, label='Califronia')
plt.xlabel("Time Since First Case (Days)")
plt.ylabel("Fraction of Population")
plt.title("Infected Cases")
plt.legend()
```

```python
fig = plt.figure(figsize=(25,15))
plt.rcParams['font.size'] = '16'
plt.plot(df_state1.index, df_state1['deaths']/1362359,label='Maine')
plt.plot(df_state2.index, df_state2['deaths']/7766925,label='Washinton')
plt.plot(df_state3.index, df_state3['deaths']/29527941,label='Texas')
plt.plot(df_state4.index, df_state4['deaths']/7766925,label='Nebraska')
plt.plot(df_state5.index, df_state5['deaths']/39664128, label='Califronia')
plt.xlabel("Time Since First Case (Days)")
plt.ylabel("Fraction of Population")
plt.title("Number of Deaths")
plt.legend()
```
